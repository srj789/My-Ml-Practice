```python

import numpy as np
import pandas as pd
dataset = pd.read_csv("C:/Users/Suraj  kumar/Desktop/ML Club/PCE-CS-III-ML-Club-master/Data Sets/Assignment5.csv" 
                      ,names = ['employee', 'works_under'])
dataset
```

```python
y = dict(zip(dataset.employee,dataset.works_under))
y
```

```python
res = []
for i in y :
    t = []
    t.append(i)
    temp = i
    while temp != y[temp] :
        temp = y[temp]
        t.append(temp)
    res.append(t)
```
